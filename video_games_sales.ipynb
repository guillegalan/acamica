```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Datasets/Video_Games_Sales_as_at_22_Dec_2016.csv')
```

```python
plt_year_mean = df.groupby(df.Platform).Year.mean()
```

```python
plt_year_mean
```

```python
df = df.join(plt_year_mean, on='Platform', lsuffix='_ori', rsuffix='_new')
```

```python
df.head(5)
```

```python
df['Year_calc'] = df.Year_ori.fillna(df.Year_new)
```

```python
df[df.Year_ori.isnull()]
```

```python
df.Year_calc = df.Year_calc.apply(np.int64)
```

```python
df.drop(columns=['Year_ori','Year_new'],inplace=True)
```

```python
df[(df.Year_calc >= 2016)]
```

```python
df[((df.NA_Sales + df.EU_Sales + df.JP_Sales) <= 0.01)]
```

```python
df = df[((df.NA_Sales + df.EU_Sales + df.JP_Sales) > 0.01)]
```

```python
df_Genre = pd.DataFrame({'cantidad' : df.groupby(["Genre"]).size()}).reset_index()
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (12,8)

generos = df_Genre.Genre
generos_cant = df_Genre.cantidad
fig1, ax1 = plt.subplots()
ax1.pie(generos_cant, labels = generos, autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```

```python
df_Genre_Year_NA = pd.DataFrame({'Ventas' : df.groupby(["Genre","Year_calc"]).NA_Sales.sum()}).reset_index()
```

```python
generos = ('Action','Sports','Misc','Role-Playing','Shooter', 'Strategy')
sns.set(style="darkgrid")
sns.lineplot(x="Year_calc", y="Ventas",
             hue="Genre",
             data=df_Genre_Year_NA[df_Genre_Year_NA.Genre.isin(generos)])
```

```python
df_Genre_Year_EU = pd.DataFrame({'Ventas' : df.groupby(["Genre","Year_calc"]).EU_Sales.sum()}).reset_index()
```

```python
generos = ('Action','Sports','Misc','Role-Playing','Shooter', 'Strategy')
sns.set(style="darkgrid")
sns.lineplot(x="Year_calc", y="Ventas",
             hue="Genre",
             data=df_Genre_Year_EU[df_Genre_Year_EU.Genre.isin(generos)])
```

```python
df_Genre_Year_JP = pd.DataFrame({'Ventas' : df.groupby(["Genre","Year_calc"]).JP_Sales.sum()}).reset_index()
```

```python
generos = ('Action','Sports','Misc','Role-Playing','Shooter', 'Strategy')
sns.set(style="darkgrid")
sns.lineplot(x="Year_calc", y="Ventas",
             hue="Genre",
             data=df_Genre_Year_JP[df_Genre_Year_JP.Genre.isin(generos)])
```
